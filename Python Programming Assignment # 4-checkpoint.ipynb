```python
# Create a list of favorite fruits
fruits = ["apple", "banana", "cherry", "date", "elderberry", "fig", "grape"]

# Indexing operation: access the first fruit
print("First fruit:", fruits[0])

# Indexing operation: access the last fruit
print("Last fruit:", fruits[-1])

# Slicing operation: access the first three fruits
print("First three fruits:", fruits[:3])

# Slicing operation: access the fruits from index 2 to 5
print("Fruits from index 2 to 5:", fruits[2:5])

# Slicing operation: access the fruits from index 3 to the end
print("Fruits from index 3 to the end:", fruits[3:])

# Slicing operation: access every other fruit
print("Every other fruit:", fruits[::2])

# Slicing operation: access the fruits in reverse order
print("Fruits in reverse order:", fruits[::-1])
```
